# Durg Classification

## To predict which type of drug should be prescribed to the patient.

### Feature engineering

```python
import pandas as pd
import numpy as np
```

```python
data_frame=pd.read_csv('Drug200.csv')
```

```python
data_frame.head()
```

```python
#coverting categorical features to numerical features.
for cols in data_frame.columns:
    if data_frame[cols].dtype=='O':
        rank=data_frame[cols].value_counts(ascending=True).index
        mapping={i:k for k,i in enumerate(rank,0)}
        data_frame[cols]=data_frame[cols].map(mapping)
```

```python
data_frame.head()
```

```python
#covert the values of 'Na_to_K' feature to log values.
data_frame['Na_to_K']=np.log(data_frame['Na_to_K'])
```

```python
data_frame.head()
```

```python
#scaling the values using MinMaxScaler.
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
data=data_frame.drop(['Drug'],axis=1)
scaler.fit(data)
```

```python
scaler.transform(data)
```

```python
transformed_data=pd.DataFrame(scaler.transform(data),columns=data.columns)
```

```python
modified_data=pd.concat([transformed_data,data_frame[['Drug']]],axis=1)
```

```python
modified_data.head()
```

```python
#saving the data to csv file.
modified_data.to_csv('modified_data.csv',index=False)
```
